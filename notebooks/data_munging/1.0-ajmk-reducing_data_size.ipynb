```python
data = pd.read_csv("/Users/alexanderkell/Documents/PhD/Projects/18-battery-optimisation/data/processed/2012-2013-solar-electricity-data.csv")
data['datetime'] = pd.to_datetime(data['datetime'])
data
```

```python
data['quarter'] = data.datetime.dt.quarter
data['month'] = data.datetime.dt.month
data['year'] = data.datetime.dt.year
data['week'] = data.datetime.dt.week
data
```

```python
for customer, dat in data.groupby(["Customer", 'month', "year"]):
    print(customer)
    print(dat.month.count())
```

```python
data_2013 = data[data.year==2013]
print(data_2013.datetime.min())
print(data_2013.datetime.max())
```

```python
deltas = data_2013[(data_2013['Consumption Category']=="solar_generation") & (data_2013['Customer']==1)].sort_values("datetime")['datetime'].diff()

gaps = deltas[deltas > timedelta(days=1)]


print(f'{len(gaps)} gaps with average gap duration: {gaps.mean()}')
for i, g in gaps.iteritems():
    gap_start = data_2013['datetime'][i - 1]
    print(f'Start: {datetime.strftime(gap_start, "%Y-%m-%d")} | '
          f'Duration: {str(g.to_pytimedelta())}')
```
